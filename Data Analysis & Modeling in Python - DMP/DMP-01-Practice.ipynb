```python
end1 = datetime.date.today()
print(end1)
start1=end1-datetime.timedelta(50)
print(start1)

df = yf.download('NVDA',start = start1, end = end1, auto_adjust=False, interval='5m')

desired_columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']




df=df[[col for col in desired_columns if col in df.columns]]

print(type(df))
print(df.shape)

```
